### Borrado de valores faltantes

```python
df_copy = df
```

```python
# axis=1 Borra las columnas 
# axis=0 Borra las filas
# how="any" Borra la columna si tiene al menos un dato nulo
# how="all" Borra la columna aunque no tenga datos nulos
df_copy.dropna(axis=1, how="any")
```

```python
df_copy_two = df
```

```python
# Relleno de valores nulos con 0
df_copy_two = df_copy_two.fillna(0)
df_copy_two
```

```python
df_copy_two.isnull().sum()
```
